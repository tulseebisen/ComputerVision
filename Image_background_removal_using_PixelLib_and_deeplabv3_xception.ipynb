## Implementaion using Deeplabv3 Xception

* Here we are using image segmentation to remove the objects segmented from the
image and place them in the new background created.
* This is done by producing a mask for the image and combining it with the modified background.
* We make use of deeplabv3+ model trained on pascalvoc dataset. 
* We want to change the background of the image to white and set color’s RGB value to white which is (255,255,255).

```python
import pixellib
from pixellib.tune_bg import alter_bg

change_bg = alter_bg()
change_bg.load_pascalvoc_model("/content/drive/MyDrive/background/deeplabv3_xception_tf_dim_ordering_tf_kernels.h5")
change_bg.color_bg("/content/drive/MyDrive/background/view1.jpeg", colors = (255, 255, 255), output_image_name="colored_bg.jpg")
```
